```python
%matplotlib inline
%reload_ext autoreload
%autoreload 2

from fastai.structured import *
from fastai.column_data import *
from IPython.display import HTML
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from itertools import combinations
import holoviews as hv

hv.extension('bokeh')
np.set_printoptions(threshold=5000, edgeitems=20)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
PATH='data/'
```

```python
matches = pd.read_csv('data/results.csv', parse_dates=['date'])
print(matches.shape)
matches.head(2)
```

```python
matches = matches.drop('neutral', axis=1) #this field is not accurate, try matches[(matches['home_team']!=matches['country']).values & (matches['neutral']==False).values]
matches = matches.rename({'home_team':'team1', 'away_team':'team2', 'home_score':'team1_goal', 'away_score':'team2_goal', 'city':'venue_city', 'country':'venue_country'}, axis=1)
matches = matches.replace({'Germany DR': 'Germany', 'China': 'China PR'})
matches['venue_country'] = matches['venue_country'].replace({'French Polynesia': 'Tahiti', 'Serbia and Montenegro': 'Serbia'})

# duplicate with team1 and team2 swapped
matches_dupe = matches.copy()
for suffix in ['', '_goal']:
    matches_dupe['tmp'] = matches_dupe['team1'+suffix]
    matches_dupe['team1'+suffix] = matches_dupe['team2'+suffix]
    matches_dupe['team2'+suffix] = matches_dupe['tmp']
    matches_dupe = matches_dupe.drop('tmp', axis=1)
matches = pd.concat([matches, matches_dupe])
# matches.sort_values('date', inplace=True)
print(matches.shape)
matches.head(2)
```

```python
rankings = pd.read_csv('data/fifa_ranking.csv', parse_dates=['rank_date'])
print(rankings.shape)
rankings.head(2)
```

```python
rankings = rankings[rankings['total_points']>0]
# rankings = rankings[['rank_date', 'rank', 'country_full', 'country_abrv', 'cur_year_avg_weighted', 'two_year_ago_weighted', 'three_year_ago_weighted']]
rankings = rankings.replace({'IR Iran': 'Iran'})
rankings['weighted_points'] =  rankings['cur_year_avg_weighted'] + rankings['two_year_ago_weighted'] + rankings['three_year_ago_weighted']
rankings = rankings.set_index(['rank_date']).groupby(['country_full'], group_keys=False).resample('D').first().fillna(method='ffill').reset_index() # Ranks for each day
matches = matches.merge(rankings, left_on=['date', 'team1'], right_on=['rank_date', 'country_full'])
matches = matches.merge(rankings, left_on=['date', 'team2'], right_on=['rank_date', 'country_full'], suffixes=('_team1', '_team2'))
add_datepart(matches, 'date', drop=False)
def add_comparison_features(df):
    df['rank_diff'] = df['rank_team1'] - df['rank_team2']
    df['rank_mean'] = (df['rank_team1'] + df['rank_team2'])/2
    df['total_points_diff'] = df['total_points_team1'] - df['total_points_team2']
    df['total_points_mean'] = (df['total_points_team1'] + df['total_points_team2'])/2
    df['cur_year_avg_diff'] = df['cur_year_avg_team1'] - df['cur_year_avg_team2']
    df['cur_year_avg_mean'] = (df['cur_year_avg_team1'] + df['cur_year_avg_team2'])/2
    df['last_year_avg_diff'] = df['last_year_avg_team1'] - df['last_year_avg_team2']
    df['last_year_avg_mean'] = (df['last_year_avg_team1'] + df['last_year_avg_team2'])/2
    df['two_year_ago_avg_diff'] = df['two_year_ago_avg_team1'] - df['two_year_ago_avg_team2']
    df['two_year_ago_avg_mean'] = (df['two_year_ago_avg_team1'] + df['two_year_ago_avg_team2'])/2
    df['three_year_ago_avg_diff'] = df['three_year_ago_avg_team1'] - df['three_year_ago_avg_team2']
    df['three_year_ago_avg_mean'] = (df['three_year_ago_avg_team1'] + df['three_year_ago_avg_team2'])/2
    return
# add_comparison_features(matches)
matches['goal_diff'] = (matches['team1_goal'] - matches['team2_goal']).astype(float)
print(matches.shape)
matches.head(2)
```

```python
# cat_vars = ['team1', 'team2', 'tournament', 'Year', 'Month', 'Week', 'Day', 'Dayofweek', 'Dayofyear', 'Is_month_end', 'Is_month_start', 'Is_quarter_end', 'Is_quarter_start', 'Is_year_end', 'Is_year_start']
cat_vars    = ['team1', 'team2', 'tournament', 'Year', 'Week', 'confederation_team1', 'confederation_team2']
# contin_vars = ['rank_team1', 'cur_year_avg_weighted_team1', 'two_year_ago_weighted_team1', 'three_year_ago_weighted_team1', 'weighted_points_team1',
#                'rank_team2', 'cur_year_avg_weighted_team2', 'two_year_ago_weighted_team2', 'three_year_ago_weighted_team2', 'weighted_points_team2',
#                'rank_diff', 'rank_mean', 'point_diff', 'point_mean', 'Elapsed']
contin_vars = ['rank_team1', 'total_points_team1', 'cur_year_avg_team1', 'last_year_avg_team1', 'two_year_ago_avg_team1', 'three_year_ago_avg_team1',
               'rank_team2', 'total_points_team2', 'cur_year_avg_team2', 'last_year_avg_team2', 'two_year_ago_avg_team2', 'three_year_ago_avg_team2',
#                'rank_diff' , 'total_points_diff' , 'cur_year_avg_diff' , 'last_year_avg_diff' , 'two_year_ago_avg_diff' , 'three_year_ago_avg_diff' ,
#                'rank_mean' , 'total_points_mean' , 'cur_year_avg_mean' , 'last_year_avg_mean' , 'two_year_ago_avg_mean' , 'three_year_ago_avg_mean' ,
               'Elapsed']
dep = ['team1_goal', 'team2_goal']
matches = matches[cat_vars+contin_vars+dep]
print(matches.shape)
matches.head(2)
```

```python
# set recent observations as test set
# train_val, test = train_test_split(matches.copy(), test_size=0.2, random_state=42)
test_ratio = 0.8
train_val_size = int(len(matches) * test_ratio)
test_idx = list(range(train_val_size, len(matches)))
train_val = matches[:train_val_size ].copy()
test      = matches[ train_val_size:].copy()

train_val = train_val.sort_values(['Year','Week'], ascending=False) # so that the recent observations were used when the laerning rate is high
train_val = train_val.reset_index(drop=True)

for v in cat_vars: 
    train_val[v] = train_val[v].astype('category').cat.as_ordered()
apply_cats(test, train_val)
    
X_train_val, y_train_val, nas, mapper = proc_df(train_val.reset_index(drop=True).drop(dep, axis=1), y_fld=None, do_scale=True)
X_test,      y_test,      nas, mapper = proc_df(test     .reset_index(drop=True).drop(dep, axis=1), y_fld=None, do_scale=True, mapper=mapper, na_dict=nas)

y_train_val = train_val[['team1_goal','team2_goal']].astype(float).values
y_test      = test     [['team1_goal','team2_goal']].astype(float).values

# val_idx = np.flatnonzero((df.index<=datetime.datetime(2014,9,17)) & (df.index>=datetime.datetime(2014,8,1)))
train_ratio = 0.8
train_size = int(len(X_train_val) * train_ratio)
val_idx = list(range(0, len(train_val)-train_size)) # set the recent obesrvations in the train_val set as validation set
md = ColumnarModelData.from_data_frame(PATH, val_idx, X_train_val, y_train_val.astype(np.float32), cat_flds=cat_vars, bs=4, test_df=X_test)
```

```python
cat_sz = [(c, len(train_val[c].cat.categories)+1) for c in cat_vars]
cat_sz
```

```python
emb_szs = [(c, min(50, (c+1)//2)) for _,c in cat_sz]
emb_szs
```

```python
m = md.get_learner(emb_szs, len(X_train_val.columns)-len(cat_vars), 0.04, 2, [1000,500], [0.001,0.01], y_range=(np.min(y_train_val), np.max(y_train_val)))
m.lr_find()
m.sched.plot(100)
```

```python
def mae(y_pred, targ):
    return mean_absolute_error(targ, y_pred)
def mae_1(y_pred, targ):
    return mean_absolute_error(targ.view(-1, 2)[:, 0:1] ,y_pred[:, 0:1])
def mae_2(y_pred, targ):
    return mean_absolute_error(targ.view(-1, 2)[:, 1:2] ,y_pred[:, 1:2])
```

```python
def do_experiment(nn, dropout, lr, n_cycle, cycle_len):

    lr = lr
    m = md.get_learner(emb_szs, len(X_train_val.columns)-len(cat_vars), 0.04, 2, nn, dropout, y_range=(np.min(y_train_val), np.max(y_train_val)))

    m.fit(lr, n_cycle=n_cycle, cycle_len=cycle_len, metrics=[mae_1, mae_2])
    m.save('model_01')
    val = train_val[train_size:len(train_val)].copy()
    val['team1_goal_pred'] = m.predict(is_test=False)[:,0]
    val['team2_goal_pred'] = m.predict(is_test=False)[:,1]
    test['team1_goal_pred'] = m.predict(is_test=True)[:,0]
    test['team2_goal_pred'] = m.predict(is_test=True)[:,1]
    print(mae(test['team1_goal_pred'], test['team1_goal']), (mae(test['team2_goal_pred'], test['team2_goal'])))
    plot = hv.Scatter(val[['team1_goal','team1_goal_pred']]) + hv.Scatter(test[['team1_goal','team1_goal_pred']])+ hv.Scatter(val[['team2_goal','team2_goal_pred']]) + hv.Scatter(test[['team2_goal','team2_goal_pred']]) 

    group_stage = pd.read_csv('data/World Cup 2018 Dataset.csv')
    group_stage = group_stage[['Team', 'Group', 'First match \nagainst', 'Second match\n against', 'Third match\n against']]
    group_stage = group_stage.dropna(how='all')
    group_stage = group_stage.replace({'IRAN': 'Iran', 'Costarica': 'Costa Rica', 'Porugal': 'Portugal', 'Columbia': 'Colombia', 'Korea' : 'Korea Republic'})
    group_stage = group_stage.set_index('Team')
    world_cup_rankings = rankings.loc[(rankings['rank_date'] == rankings['rank_date'].max()) & rankings['country_full'].isin(group_stage.index.unique())]
    world_cup_rankings = world_cup_rankings.set_index(['country_full'])
    group_stage['points'] = 0
    group_stage['goal_pred_total'] = 0


    def get_goal_diff_perd():
        print("{} vs. {}: ".format(team1, team2), end='')
        row = pd.DataFrame(np.nan, index=[0], columns=X_train_val.columns)
    #     print(row)
        row['team1'] = team1
        row['team2'] = team2
        row['tournament'] = 'FIFA World Cup'
        row['date'] = '2018-06-14'
        add_datepart(row, 'date', drop=True)
        row['rank_team1'] = world_cup_rankings.loc[team1, 'rank']
        row['total_points_team1'] = world_cup_rankings.loc[team1, 'total_points']
        row['cur_year_avg_team1'] = world_cup_rankings.loc[team1, 'cur_year_avg']
        row['last_year_avg_team1'] = world_cup_rankings.loc[team1, 'last_year_avg']
        row['two_year_ago_avg_team1'] = world_cup_rankings.loc[team1, 'two_year_ago_avg']
        row['three_year_ago_avg_team1'] = world_cup_rankings.loc[team1, 'three_year_ago_avg']
        row['rank_team2'] = world_cup_rankings.loc[team2, 'rank']
        row['total_points_team2'] = world_cup_rankings.loc[team2, 'total_points']
        row['cur_year_avg_team2'] = world_cup_rankings.loc[team2, 'cur_year_avg']
        row['last_year_avg_team2'] = world_cup_rankings.loc[team1, 'last_year_avg']
        row['two_year_ago_avg_team2'] = world_cup_rankings.loc[team2, 'two_year_ago_avg']
        row['three_year_ago_avg_team2'] = world_cup_rankings.loc[team2, 'three_year_ago_avg']
        add_comparison_features(row)

        row = row[X_train_val.columns]
        apply_cats(row, train_val)
    #     for v in contin_vars:
    #         row[v] = row[v].fillna(0).astype('float32')
        row, _, _, _ = proc_df(row.reset_index(drop=True), y_fld=None, do_scale=True, mapper=mapper, na_dict=nas)

        md_row = ColumnarModelData.from_data_frame(PATH, val_idx, X_train_val, y_train_val.astype(np.float32), cat_flds=cat_vars, bs=4, test_df=row)
        team1_goal_pred = m.predict_dl(md_row.test_dl)[:,0][0]
        team2_goal_pred = m.predict_dl(md_row.test_dl)[:,1][0]
        return team1_goal_pred, team2_goal_pred

    for group in group_stage['Group'].unique().tolist():
        print('Group {}:'.format(group))
        for team1, team2 in combinations(group_stage.query('Group == "{}"'.format(group)).index, 2):     
            team1_goal_pred, team2_goal_pred = get_goal_diff_perd()
            group_stage.loc[team1, 'goal_pred_total'] += team1_goal_pred
            group_stage.loc[team2, 'goal_pred_total'] += team2_goal_pred
            points = 0
            if team1_goal_pred - team2_goal_pred < -.5:
                group_stage.loc[team2, 'points'] += 3
                print('{:.2f} : {:.2f} ({} wins)'.format(team1_goal_pred, team2_goal_pred, team2))
            if team1_goal_pred - team2_goal_pred >= -.5 and team1_goal_pred - team2_goal_pred < .5:
                points = 1
                print('{:.2f} : {:.2f} (Draw)'.format(team1_goal_pred, team2_goal_pred))
                group_stage.loc[team1, 'points'] += 1
                group_stage.loc[team2, 'points'] += 1
            if team1_goal_pred - team2_goal_pred>= .5:
                points = 3
                group_stage.loc[team1, 'points'] += 3
                print('{:.2f} : {:.2f} ({} wins)'.format(team1_goal_pred, team2_goal_pred, team1))

    group_stage.sort_values(by=['Group', 'points', 'goal_pred_total'], ascending=[True, False, False], inplace=True)
    group_stage.reset_index(inplace=True)

    round_of_16_teams = group_stage.groupby('Group').nth([0, 1]) # select the top 2
    round_of_16_teams = round_of_16_teams.reset_index()
    round_of_16_teams = round_of_16_teams.loc[[0,3,4,7,8,11,12,15,1,2,5,6,9,10,13,14]]
    round_of_16_teams = round_of_16_teams.set_index('Team')

    winners = round_of_16_teams.index
    semi_final_losers = []
    for stage in ['Round of 16', 'Quarter-finals', 'Semi-finals', 'Final', 'Thrid Place Playoff']:
        print('{}:'.format(stage))
        last_round_winners = round_of_16_teams.loc[winners]
        winners = []
        if stage == 'Thrid Place Playoff':
            no_of_matches = range(1)
        else:
            no_of_matches = range(int(len(last_round_winners)/2))
        for i in no_of_matches:
            if stage == 'Thrid Place Playoff':
                semi_final_losers  = round_of_16_teams.loc[semi_final_losers]
                team1 = semi_final_losers.index[i*2]
                team2 = semi_final_losers.index[i*2+1]
            else:
                team1 = last_round_winners.index[i*2]
                team2 = last_round_winners.index[i*2+1]
            team1_goal_pred, team2_goal_pred = get_goal_diff_perd()
            if team1_goal_pred - team2_goal_pred >= 0:
                print('{:.2f} : {:.2f} ({} wins)'.format(team1_goal_pred, team2_goal_pred, team1))
                winners.append(team1)
                if stage == 'Semi-finals':
                    semi_final_losers .append(team2)
            else:
                print('{:.2f} : {:.2f} ({} wins)'.format(team1_goal_pred, team2_goal_pred, team2))
                winners.append(team2)
                if stage == 'Semi-finals':
                    semi_final_losers .append(team1)
                    
    return plot
```
